```python
def convert1(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count !=3:
            L.append(i['name'])
            count=count+1
        else:
            break
    return L
            
```

```python
# fetching movie director
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
# final_movie['tags']=final_movie['tags'].apply(stem)
```

```python
final_movie['tags'] = final_movie['tags'].apply(lambda x:" ".join(x))
```

```python
final_movie['tags']=final_movie['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer (max_features=5000, stop_words='english')
```

```python
vector=cv.fit_transform(final_movie['tags']).toarray()
```

```python
vector.shape
```

```python
vector[0]
```

```python
cv.get_feature_names() # gives all the features words
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend (movie):
    movie_index = final_movie[final_movie['title'] == movie].index [0]
    distances = similarity [movie_index]
    movies_list = sorted (list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(final_movie.iloc[i[0]].title)
```

```python
recommend('X-Men: The Last Stand')
```
