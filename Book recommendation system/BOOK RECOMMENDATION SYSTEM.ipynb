# Now,POPULARITY BASED RECOMMENDATION SYSTEM

```python
rating=ratings.merge(books,on='ISBN')
```

```python
rating
```

```python
num_ratings=rating.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_ratings.columns = ["Book-Title","num of ratings"]
```

```python
num_ratings
```

```python
avg_rating=rating.groupby('Book-Title').mean()['Book-Rating'].reset_index()
```

```python
avg_rating.columns = ["Book-Title","Average of ratings"]
avg_rating
```

```python
popularity_df= num_ratings.merge(avg_rating,on='Book-Title')
```

```python
popularity_df
```

```python
popular=popularity_df[popularity_df['num of ratings']>250].sort_values('Average of ratings',ascending=False).head(50)
```

```python
popular
```

```python
popular=popular.merge(books, on='Book-Title').drop_duplicates('Book-Title')
```

```python
popular=popular[['Book-Title','num of ratings','Average of ratings','Book-Author','Publisher','Image-URL-M']]
```

```python
popular
```
